```python
merged1_df = Data1_df.merge(
    Key1_df[['Blind ID','Genotype','Density']],
    on='Blind ID',
    how='left'
).drop(columns=['Blind ID'])

merged2_df = Data2_df.merge(
    Key2_df[['Blind ID','Genotype','Density']],
    on='Blind ID',
    how='left'
).drop(columns=['Blind ID']) 

#merges the Blind ID to the porper genotype and density 

print(merged1_df)
print(merged2_df)
```

```python
merged1_df = (
    merged1_df
      .rename(columns={'Density_x': 'Density'}) 
      .drop(columns=['Density_y']))  

merged2_df = (
    merged2_df
      .rename(columns={'Density_x': 'Density'})  
      .drop(columns=['Density_y']))  
#renames the Density column to provide more clarity (both density columns were necessary for merging to ensure the Genotype is associated to the right Density and Blind ID).

print(merged1_df)
print(merged2_df) 
```

```python
def clean_data(df):
    
    df = df[df['Trial Day'] >= 3].copy() 
    #creates a copy of the data frame focusing on days after 2

    avg_pupation = (
        df[df['Pupae Emerged'] > 0]
        .groupby(['Genotype', 'Density'])['Trial Day']
        .mean()
        .unstack()) 
    #allows me to locate when pupae first appear

    avg_emergence = (
        df[df['Mosquitoes Emerged'] > 0]
        .groupby(['Genotype', 'Density'])['Trial Day']
        .mean()
        .unstack()) 
    #allows me to locate when mosquitoes first appear

    total_dead     = df.groupby(['Genotype', 'Density'])['# of Deceased'].sum()
    start_density  = df.groupby(['Genotype', 'Density'])['Density'].first()
    avg_mortality  = (total_dead / start_density).unstack() 
    #calculates the mortality rate by finding the sum and dividing them by the original densities

    total_males    = df.groupby(['Genotype', 'Density'])['Males Emerged'].sum()
    total_females  = df.groupby(['Genotype', 'Density'])['Females Emerged'].sum()
    avg_sex_ratio  = (total_males / total_females).unstack() 
    # calculates the sex ratio by finding the sum of males and females individually and dividing males by females

    return avg_pupation, avg_emergence, avg_mortality, avg_sex_ratio
```

```python
for df, label in [(merged1_df, 'Trial 1'), (merged2_df, 'Trial 2')]:  
    avg_pupation, avg_emergence, avg_mortality, avg_sex_ratio = clean_data(df)  
    #creates a loop to run the following graph codes trhough trial 1 and trial 2 

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    avg_pupation.plot.bar(ax=axes[0,0], rot=1)
    axes[0,0].set(
        title=f'{label} â€“ Avg. Pupation Day',
        ylabel='Day') 
    #bar graph for average pupation in the top left

    avg_emergence.plot.bar(ax=axes[0,1], rot=1)
    axes[0,1].set(
        title=f'{label} â€“ Avg. Emergence Day',
        ylabel='Day') 
    #bar graph for average mosquito emergence (no gender included) in top right

    avg_mortality.plot.bar(ax=axes[1,0], rot=1)
    axes[1,0].set(
        title=f'{label} â€“ Mortality Rate',
        ylabel='Fraction Deceased') 
    #bar graph for mortality rate in bottom left

    avg_sex_ratio.plot.bar(ax=axes[1,1], rot=1)
    axes[1,1].set(
        title=f'{label} â€“ Sex Ratio (M:F)',
        ylabel='Ratio') 
    #bar graph for sex ration M to F in bottom right

    plt.tight_layout()
    plt.show()
    #load relevant graphs to conclude the fitness cost of carrying the pesticide resistant genotypes at high(100) and low(25) densities
```
